행렬 연산을 이용하여 OLS 추정량, t-value, p-value, R² 직접 구해보기

- DGP : yi = β0 + β2edu + β4exper + β3tenure + ei, ei ~ i. i. d. N(0, σ²)
- β_hat = (X_T@X)^-1 @ X_T @ y
- Var_hat(β_hat) = σ_hat^2(X_T@X)^-1
- smf.ols()를 사용한 결과와 일치하는 것을 반드시 확인

```python
# X
wage1['intercept'] = 1
X = wage1[['intercept', 'educ', 'exper', 'tenure']]
# y
y = wage1['wage']
# β_hat
beta_hat = (np.linalg.inv(X.T@X)) @ (X.T) @ (y)
beta_hat.index = ['Intercept', 'educ', 'exper', 'tenure']
beta_hat.name = 'b'

Table1 = pd. DataFrame({ 'b': round(beta_hat, 4)}) 
print(f'행렬연산: \n{Table1}\n')

# smf.ols()
model = smf.ols(formula='wage ~ educ + exper + tenure', data=wage1)
results = model.fit()

Table2 = pd. DataFrame({ 'b': round(results.params, 4)}) 
print(f'OLS추정: \n{Table2}\n')
```

```python
# e
n = len(X)
y_hat = X.values @ beta_hat.values
e_hat = y - y_hat
sse = sum(e_hat**2)
sig = np.sqrt(sse/(n-4))
cov = (sig**2) * (np.linalg.inv(X.T@X))

# t-value
t_values = pd.Series()
for num, name in enumerate(beta_hat.index):
    t_values[name] = (beta_hat[num]-0)/np.sqrt(cov[num,num])

Table1 = pd. DataFrame({ 't': round(t_values, 4)}) 
print(f'행렬연산: \n{Table1}\n')

# smf.ols()
model = smf.ols(formula='wage ~ educ + exper + tenure', data=wage1)
results = model.fit()

Table2 = pd. DataFrame({ 't': round(results.tvalues, 4)}) 
print(f'OLS추정: \n{Table2}\n')
```

```python
# p-value 
p_values = pd.Series()
for num, name in enumerate(beta_hat.index):
    p_val_two = 2*(1 - stats.t.cdf(np.abs(t_values[num]), n-2))
    p_values[name] = p_val_two
Table1 = pd. DataFrame({ 'pval': round(p_values, 4)}) 
print(f'행렬연산: \n{Table1}\n')

# smf.ols()
model = smf.ols(formula='wage ~ educ + exper + tenure', data=wage1)
results = model.fit()

Table2 = pd. DataFrame({ 'pval': round(results.pvalues, 4)}) 
print(f'OLS추정: \n{Table2}\n')
```

```python
# R^2
y_hat = X.values @ beta_hat.values
y_bar = y.mean()
y = y

SST = sum((y - y_bar ) **2)      # 다른 이름: TSS
SSR = sum((y_hat - y_bar ) **2)  # 다른 이름: ESS
SSE = sum( (y - y_hat)**2)       # 다른 이름: RSS
          
R2 = 1 - SSE/SST
print("행렬연산(1-SSE/SST)")
print("R2: ", round(R2,4))

R2 = SSR/SST
print("\n행렬연산(SSR/SST)")
print("R2: ", round(R2,4)) # 같은지 확인

# smf.ols()
model = smf.ols(formula='wage ~ educ + exper + tenure', data=wage1)
results = model.fit()
print("\nOLS추정")
print(f'R2:  {round(results.rsquared, 4) }\n')
```
