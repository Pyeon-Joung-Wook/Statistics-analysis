```python
def get_finance_data(path) : 
    data_path = path
    raw_data = pd.read_excel(data_path, index_col=0)
    big_col = list(raw_data.columns)
    small_col = list(raw_data.iloc[0])
    
    new_big_col = []
    for num, col in enumerate(big_col) : 
        if num == 0 : 
            new_big_col.append(big_col[num])
        else : 
            if 'Unnamed' in col : 
                new_big_col.append(new_big_col[num-1])
            else : 
                new_big_col.append(big_col[num])
                
    raw_data.columns = [new_big_col, small_col]
    clean_df = raw_data.loc[raw_data.index.dropna()]
    try : 
        clean_df = clean_df.drop('IFRS(연결)')
    except KeyError:
        pass
        
    return clean_df
```

NaN이 숫자로 취급되지 않기에 sorting 중 에러가 날 경우가 있다. 그럴 경우 아래와 같이 to_numeric 메소드를 통해 데이터를 숫자형으로 바꿔 줄 수 있다.

```python
invest_df[('2015/12', 'PER')] = pd.to_numeric(invest_df[('2015/12', 'PER')])
per_sorted = invest_df.sort_values(by=('2015/12', 'PER'))
per_sorted.head(10)
```

ROA 컬럼의 경우에는 'N/A(IFRS)'와 같은 문자 데이터가 있는 데, 위 코드만으로는 숫자형 데이터로 바꿀 수 없다.

따라서 일일이 NaN으로 변경해주어야 한다.

```python
import numpy as np

def check_IFRS(x) : 
    
    if x == 'N/A(IFRS)' : 
        return np.NaN
        # 기존에 있던 값 대신 NaN으로 반환
    
    else : 
        return x
    
fr_df[('2015/12', 'ROA')] = fr_df[('2015/12', 'ROA')].apply(check_IFRS)
```

```python
fr_df[('2015/12', 'ROA')] = pd.to_numeric(fr_df[('2015/12', 'ROA')])
sorted_roa = fr_df.sort_values(by = ('2015/12', 'ROA'), ascending = False)
sorted_roa.head(10)
```

```python
def low_per(invest_df, index_date, num) : 
    invest_df[(index_date, 'PER')] = pd.to_numeric(invest_df[index_date, 'PER'])
    per_sorted = invest_df.sort_values(by=(index_date, 'PER'))
    return per_sorted[index_date][:num]
```

```python
low_per(invest_df, '2018/12', 10)
```

```python
def high_roa(fr_df, index_date, num) : 
    fr_df[(index_date, 'ROA')] = fr_df[(index_date, 'ROA')].apply(check_IFRS)
    fr_df[(index_date, 'ROA')] = pd.to_numeric(fr_df[(index_date, 'ROA')])
    sorted_roa = fr_df.sort_values(by = (index_date, 'ROA'), ascending = False)
    return sorted_roa[index_date][:num]
```

```python
high_roa(fr_df, '2018/12', 10)
```

# 마법공식

영업이익(EBIT) 대비 기업가치(EV)가 낮고, 자본수익률(ROC·영업이익/투자자본)이 높은 20~30개 종목에 투자하는 방법

```python
date = '2018/12'

magic_per = low_per(invest_df, date, None)
magic_roa = high_roa(fr_df, date, None)
```

```python
magic_per['PER 순위'] = magic_per['PER'].rank()
magic_roa['ROA 순위'] = magic_roa['ROA'].rank(ascending = False)
display(magic_per.head())
display(magic_roa.head())
```

```python
(magic_per['PER 순위'] + magic_roa['ROA 순위']).sort_values()
```

```python
(magic_per['PER 순위'] + magic_roa['ROA 순위']).rank().sort_values()
```

※ 주의 : 둘의 합산 순위가 같은 경우 rank를 매기면 같은 등수가 나오게 되는 문제가 발생

```python
magic = pd.merge(magic_per, magic_roa, how = 'outer', left_index = True, right_index = True)
magic['마법공식 순위'] = (magic['PER 순위'] + magic['ROA 순위']).rank().sort_values()
magic = magic.sort_values(by = '마법공식 순위')

magic.head(10)
```

```python
def magic_formula(fr_df, invest_df, date, num) : 
    
    magic_per = low_per(invest_df, date, None)
    magic_roa = high_roa(fr_df, date, None)
    magic_per['PER 순위'] = magic_per['PER'].rank()
    magic_roa['ROA 순위'] = magic_roa['ROA'].rank(ascending = False)
    
    magic = pd.merge(magic_per, magic_roa, how = 'outer', left_index = True, right_index = True)
    magic['마법공식 순위'] = (magic['PER 순위'] + magic['ROA 순위']).rank().sort_values()
    magic = magic.sort_values(by = '마법공식 순위')

    return magic[:num]
```

```python
magic_formula(fr_df, invest_df, '2018/12', 10)
```

# 저평가 주

```python
def get_value_rank(invest_df, value_type, index_date, num) : 
    
    invest_df[(index_date, value_type)] = pd.to_numeric(invest_df[(index_date, value_type)])
    value_sorted = invest_df.sort_values(by=(index_date, value_type), ascending = True)[index_date]
    value_sorted[value_type + '순위'] = value_sorted[value_type].rank()
    
    return value_sorted[[value_type, value_type + '순위']][:num]
```

```python
get_value_rank(invest_df, 'PSR', '2017/12', 10)
```

```python
date = '2018/12'
per = get_value_rank(invest_df, 'PER', '2018/12', None)
pbr = get_value_rank(invest_df, 'PBR', '2018/12', None)
value_combo = pd.merge(per, pbr, how = 'outer', left_index = True, right_index = True)
value_combo['종합순위'] = (value_combo['PER순위'] + value_combo['PBR순위']).rank()
value_combo = value_combo.sort_values(by = '종합순위')
value_combo[:20]
```

```python
def make_value_combo(value_list, invest_df, index_date, num) : 
    
    for i, value in enumerate(value_list) : 
        temp_df = get_value_rank(invest_df, value,  index_date, None) 
        
        if i == 0 :
            value_combo_df = temp_df
            rank_combo = temp_df[value + '순위']
        else : 
            value_combo_df = pd.merge(value_combo_df, temp_df, how = 'outer', right_index = True, left_index = True)
            rank_combo = rank_combo + temp_df[value + '순위']
    
    value_combo_df['종합순위'] = rank_combo.rank()
    value_combo_df = value_combo_df.sort_values(by = '종합순위')
    
    return value_combo_df[:num]
```

# F - Score

F-Score는 총 9가지 지표가 있는데 수익성, 재무건전성, 효율성을 평가하는 지표입니다. 

   ------------------------------------------------------------------------------------------------------------------   

1. 수익성 (Profitability) 

    (1) 자산수익률(ROA;Net income before extraordinary items/총자산) : 0보다 크면 1, 아니면 0 

    -> ROA가 0보다 크면 1, 아니면 0 

    (2) 영업현금흐름(CFO;영업현금흐름/총자산): 0보다 크면 1, 아니면 0 

    -> 영업현금흐름이 0보다 크면 1, 아니면 0 

    (3) Δ ROA(올해 ROA/전년 ROA): 1보다 크면 1, 아니면 0 

    -> 전년동기 대비 ROA가 증가했으면 1, 아니면 0 

    (4) CFO-ROA : 0보다 크면 1, 아니면 0 

    -> 영업현금흐름이 자산수익보다 크면 1, 아니면 0 

------------------------------------------------------------------------------------------------------------------   
       

2. 재무건전성(Financial performance signals) 

    (1) Δ Leverage(올해 장기부채-전년 장기부채) : 0보다 크면 1, 아니면 0 

    -> 장기부채비율이 감소했으면 1, 아니면 0 

    (2) Δ Liquid(올해 유동비율-전년 유동비율) : 0보다 크면 1, 아니면 0 

    -> 유동비율이 증가했으면 1, 아니면 0 

    (3) EQISS(신주발행) : 신주 미발행 1, 아니면 0 

    -> 신주발행 없으면 1, 아니면 0 
------------------------------------------------------------------------------------------------------------------   
   

3. 영업효율성(Operating efficiency) 

    (1) ΔMARGIN (올해 매출총이익 - 전년 매출총이익) : 0보다 크면 1, 아니면 0 

    -> 매출총이익 증가했으면 1, 아니면 0 

    (2) ΔTURN (올해 총자산회전율 - 전년 총자산회전율) : 0보다 크면 1, 아니면 0 

    -> 총자산회전율이 증가했으면 1, 아니면 0 
------------------------------------------------------------------------------------------------------------------   
   
총 9가지의 지표이지만 정리하면 아래와 같습니다. 

F-Score는 전년동기대비 수익 증가(ROA, CFO, 매출총이익), 영업효율성증가(총자산회전율), 유동비율증가(유동비율), 부채감소(장기부채비율)한 기업 중에서 실제로 현금 유입이 있고 신주발생이 없는(즉, 주당순이익 희석이 없음.) 기업을 선정하는 지표입니다. 

출처 - https://blog.naver.com/mymoneymoney/221232481172

```python
date = '2018/12'
fsscore_df = fs_df[date]
fscore_df['당기순이익 점수'] = fscore_df['당기순이익'] > 0
fscore_df['영업현금흐름 점수'] = fscore_df['영업활동으로인한현금흐름'] > 0
fscore_df['순이익 대비 영업현금흐름 점수'] = fscore_df['영업활동으로인한현금흐름'] >  fscore_df['당기순이익']
fscore_df.head()
```

```python
fsscore_df[['당기순이익 점수', '영업현금흐름 점수', '순이익 대비 영업현금흐름 점수']].sum(axis=1)
```

```python
def get_fscore(fs_df, index_date, num) : 
    fscore_df = fs_df[index_date]
    fscore_df['당기순이익 점수'] = fscore_df['당기순이익'] > 0
    fscore_df['영업현금흐름 점수'] = fscore_df['영업활동으로인한현금흐름'] > 0
    fscore_df['순이익 대비 영업현금흐름 점수'] = fscore_df['영업활동으로인한현금흐름'] >  fscore_df['당기순이익']
    fscore_df['종합점수'] = fscore_df[['당기순이익 점수', '영업현금흐름 점수', '순이익 대비 영업현금흐름 점수']].sum(axis=1)
    fscore_df = fscore_df[fscore_df['종합점수'] == 3]
    return fscore_df[:num]
```
